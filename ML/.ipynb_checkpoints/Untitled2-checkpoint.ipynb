```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import re

data = {
    'Vehicle Location': [
        'POINT (-122.206146 47.839957)', 'POINT (-122.30207 47.64085)', 'POINT (-122.30207 47.64085)',
        'POINT (-122.2107298 47.7977268)', 'POINT (-122.2026532 47.7210518)', 'POINT (-122.5730403 47.4282964)',
        'POINT (-122.30207 47.64085)', 'POINT (-122.2453 47.499708)', 'POINT (-122.342434 47.736655)',
        'POINT (-122.1819876 47.5098889)', 'POINT (-122.13158 47.67858)', 'POINT (-122.3128637 47.6104992)',
        'POINT (-122.5305071 47.6400595)', 'POINT (-122.7862046 48.1129794)', 'POINT (-122.30253 47.72656)',
        'POINT (-122.0430602 47.6259676)', 'POINT (-122.3581348 47.3097552)', 'POINT (-122.1197677 47.6103472)',
        'POINT (-122.377938 47.690166)', 'POINT (-122.1872 47.61001)', 'POINT (-122.304212 47.664294)',
        'POINT (-122.1550193 47.2884536)', 'POINT (-122.3499053 47.673887)'
    ]
}

df = pd.DataFrame(data)

df['Latitude'] = df['Vehicle Location'].apply(lambda x: float(x.split()[2].strip(')')))
df['Longitude'] = df['Vehicle Location'].apply(lambda x: float(x.split()[1].strip('(')))

m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)
heat_data = [[row['Latitude'], row['Longitude']] for index, row in df.iterrows()]
HeatMap(heat_data).add_to(m)

m
```
