```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# Load the dataset from a CSV file
missing_values = ["N/A", "na", "--", "NaN", "NULL"]
df = pd.read_csv("Electric_Vehicle_Population_Data.csv", na_values=missing_values)

print(df)

# Extract Latitude and Longitude from the 'Vehicle Location' column, handling NaN values
df['Latitude'] = df['Vehicle Location'].apply(lambda x: float(x.split()[2].strip(')')) if pd.notna(x) else None)
df['Longitude'] = df['Vehicle Location'].apply(lambda x: float(x.split()[1].strip('(')) if pd.notna(x) else None)

print(df['Longitude'])

# Create a folium map centered at an average location, ignoring NaN values
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)

# Add markers for each vehicle location, skipping NaN entries
for _, row in df.iterrows():
    if pd.notna(row['Latitude']) and pd.notna(row['Longitude']):
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=f"Vehicle Location: {row['Latitude']}, {row['Longitude']}",
            icon=folium.Icon(color='blue', icon='car', prefix='fa')
        ).add_to(m)

# Optional: Add a heat map layer to show density, ignoring NaN entries
heat_data = [[row['Latitude'], row['Longitude']] for _, row in df.iterrows() if pd.notna(row['Latitude']) and pd.notna(row['Longitude'])]
HeatMap(heat_data).add_to(m)

# Save map to an HTML file and display it
m.save('ev_location_map.html')
```
