```python
import pandas as pd
import folium
from folium.plugins import HeatMap

def plot_ev_locations(file_path):
    df = pd.read_csv(file_path)

    if 'Vehicle Location' in df.columns:
        map_center = [37.0902, -95.7129]
        m = folium.Map(location=map_center, zoom_start=5)
        
        for _, row in df.iterrows():
            folium.Marker(
                location=[row['Latitude'], row['Longitude']],
                popup=f"{row['Vehicle Location']}: {row['Electric Range']} EVs",
                icon=folium.Icon(color='blue', icon='car', prefix='fa')
            ).add_to(m)
        
        heat_data = [[row['Latitude'], row['Longitude'], row['Electric Range']] for index, row in df.iterrows()]
        HeatMap(heat_data).add_to(m)
        
        m.save('ev_spatial_distribution_map.html')
        print("Map created and saved as ev_spatial_distribution_map.html")
    else:
        print("The file does not contain a 'Vehicle Location' column")

file_path = 'Electric_Vehicle_Population_Data.csv'
plot_ev_locations(file_path)
```
